# Question 1. Scrape  the  
details  of  
most  viewed  videos  on  
YouTube  
from  Wikipedia.  
Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:  A) 
Rank   
B) Name   
C) Artist   
D) Upload date   
E) Views   

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
View=[]
```

```python
s_name=driver.find_elements(By.XPATH,"//table[@class='sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter']/tbody/tr/td[1]")
for i in s_name:
    names=(i.text)
    Name.append(names)
    
s_artist=driver.find_elements(By.XPATH,"//table[@class='sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter']/tbody/tr/td[2]")
for i in s_artist:
    artists=(i.text)
    Artist.append(artists)
    
u_date=driver.find_elements(By.XPATH,"//table[@class='sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter']/tbody/tr/td[4]")
for i in u_date:
    date=(i.text)
    Upload_date.append(date)
    
s_view=driver.find_elements(By.XPATH,"//table[@class='sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter']/tbody/tr/td[3]")
for i in s_view:
    views=(i.text)
    View.append(views)
```

```python
Artist
```

```python
Upload_date
```

```python
View
```

```python
len(Name),len(Artist),len(Upload_date),len(View)
```

```python
df=pd.DataFrame({'Name':Name,'Artist':Artist,'Upload Date':Upload_date,'View':View})
df
```

```python
driver.close()
```

# Question 5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/  You have to find the 
following details:   
A) Song name   
B) Artist name   
C) Last week rank   
D) Peak rank   
E) Weeks on board   
Note: - From the home page you have to click on the charts option then hot 100-page link through code.   

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.get('https://www.billboard.com/')
```

```python
charts=driver.find_element(By.XPATH,"/html/body/div[3]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a")
charts.click()
```

```python
view_charts=driver.find_element(By.XPATH,"/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[3]/a")
view_charts.click()
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.action_chains import ActionChains
```
