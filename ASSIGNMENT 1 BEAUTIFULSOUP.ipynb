```python
from bs4 import BeautifulSoup
import requests
```

```python
# question 1
# send get request to the webpage server to get code
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page

#get page content
soup=BeautifulSoup(page.content)

#get all header tags 

header_tags=[]
for i in soup.find_all(['h1','h2','h3','h4','h5','h6']):
    header_tags.append(i.text)
    
#make dataframe 

import pandas as pd 
df=pd.DataFrame({'Header Tags':header_tags})
df
```

```python
#question2
#send a request to the webpage server
page=requests.get('https://presidentofindia.nic.in/former-presidents')

#get a page content
soup=BeautifulSoup(page.content)

#get a former president of india
former_president=[]
for i in soup.find_all(["h3"]):
    former_president.append(i.text)
    
#get a term of office
term=[]
for i in soup.find_all(["h5"]):
    term.append(i.text)
    
#make a DataFrame 
import pandas as pd 
df=pd.DataFrame({'Name':former_president,'Term of Office':term})
df
```

```python
#question3(a)
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')

#page content
Soup=BeautifulSoup(page.content)

#scrap the table 
team_data=[]
table = Soup.find("table",class_="table")
row = table.find_all("tr")
for row in row[1:11]:
    cells=row.find_all("td")
    team=cells[1].text.strip()
    matches=cells[2].text.strip()
    points=cells[3].text.strip()
    rating=cells[4].text.strip()
    team_data.append([team,matches,points,rating])

# make a Dataframe
import pandas as pd
df=pd.DataFrame(team_data,columns=["Teams","Matches","Points","Rating"])
df
```

```python
#question6
page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')

soup=BeautifulSoup(page.content)

paper_title=[]
for i in soup.find_all(['h2']):
    paper_title.append(i.text)
    
author=[]
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(i.text)
    
published_date=[]
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    published_date.append(i.text)

import pandas as pd
df=pd.DataFrame({"Title":paper_title,"Author":author,"Publisher Date":published_date})
df
```

```python
#question 7
#send get request to the webpage server to get the source code of the page

page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

#page content
Soup = BeautifulSoup(page.content)

#scraping resturant name 
resturant_name=[]
for i in Soup.find_all('div',class_="restnt-info cursor"):
    resturant_name.append(i.text)
    
#scraping cuisine
cuisine=[]
for i in Soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
    
#scraping location
location =[]
for i in Soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
#scraping ratings
ratings=[]
for i in Soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
    
#scraping image
images=[]
for i in Soup.find_all("img", class_="no-img"):
    images.append(i['data-src'])
    
#make a dataframe 
import pandas as pd
df= pd.DataFrame({'Resturant Name':resturant_name,'Cuisine':cuisine,'Location':location,'Rating':ratings,'Image':images})
df
```
