# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.get('https://www.shine.com/')
```

```python
desgination=driver.find_element(By.ID,"id_q")
desgination.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.ID,"id_loc")
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.XPATH,"//button[@class=' btn btn-secondary undefined']")
search.click()
```

```python
Title=[]
Location=[]
Company_name=[]
Experience_req=[]
```

```python
for i in range(2):
    title=driver.find_elements(By.XPATH,"//h2[@itemprop='name']/a")
    location=driver.find_elements(By.XPATH,"//div[@class=' jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2']")
    company_name=driver.find_elements(By.XPATH,"//div[@class='jobCard_jobCard_cName__mYnow']")
    experience_req=driver.find_elements(By.XPATH,"//div[@class=' jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t']")
    
    
    for j in title:
        Title.append(j.text)
    Title[:10]
    
    
    for k in location:
        Location.append(k.text)
    Location[:10]
    
    
    for l in company_name:
        Company_name.append(l.text)
    Company_name[:10]
    
    
    for m in experience_req:
        Experience_req.append(m.text)
    Experience_req[:10]
```

# Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,
Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome()
driver.get('http://www.jagranjosh.com/')
```

```python
gk_option=driver.find_element(By.LINK_TEXT,"GK")
gk_option.click()
```

```python
pm_option=driver.find_element(By.LINK_TEXT,"List of Prime Ministers of India")
pm_option.click()
```

```python
data=[]
table=driver.find_elements(By.XPATH,"//table[@class='table4']")
rows=table.find_elements(By.TAG_NAME,'tr')
for row in rows:
    cols = row.find_elements(By.TAG_NAME,'td')
    if len(cols) == 4:
        name = cols[0].text
        born_dead = cols[1].text
        term_of_office = cols[2].text
        remarks = cols[3].text
        data.append([name,born_dead, term_of_office, remarks])
```

```python
df=pd.DataFrame(data, columns=['Name','Born-Dead','Term of Office','Remarks'])
df
```
